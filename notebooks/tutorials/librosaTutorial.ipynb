```python
import librosa
import librosa.display
from IPython.display import Audio, display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from glob import glob

from itertools import cycle

sns. set_theme (style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

## loading the audio files paths from the dataset

```python
audio_files = glob("../data/raw/dev-clean/**/*.flac", recursive=True)
```

## playing one of the audio files

```python
print(audio_files[0])

y, sr = librosa.load(
    audio_files[0],
    sr=None
)

display(Audio(y, rate=sr))
```

### reading the audio file 

```python
y, sr = librosa.load(audio_files[0])
print(y[:10])  # print first 10 samples
print(f"Audio shape: {y.shape}, Audio type: {y.dtype}")
print(f"Sample rate: {sr}, audio duration: {len(y)/sr:.2f} seconds")
```

## ploting the graph

```python
pd.Series(y).plot(figsize=(15, 5))
```

## trim and slint part of the audio at the starting and at the end

```python
y_trimed, _ = librosa.effects.trim(y)
pd.Series(y_trimed).plot(figsize=(15, 5))
```
