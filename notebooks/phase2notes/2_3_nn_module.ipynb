## **📚 THEORY: What is nn.Module?**

### **The Big Idea**

`nn.Module` is PyTorch's **building block** for neural networks. It:
- Tracks parameters automatically
- Handles forward pass
- Works with optimizers seamlessly
- Enables easy model saving/loading

### **Two Key Methods**

Every `nn.Module` has:

**1. `__init__()` - Define your layers**
```python
def __init__(self):
    super().__init__()  # ALWAYS call this first!
    self.layer1 = nn.Linear(5, 64)  # Define layers
    self.layer2 = nn.Linear(64, 3)
```

**2. `forward()` - Define computation**
```python
def forward(self, x):
    x = self.layer1(x)      # Pass through layer 1
    x = torch.relu(x)       # Apply activation
    x = self.layer2(x)      # Pass through layer 2
    return x                # Return output
```

## **🔄 COMPARISON: Manual vs nn.Module**

### **Your Session 2.2 Code (Manual):**

```python
# Define weights manually
W1 = torch.randn(64, 5, requires_grad=True)
b1 = torch.randn(64, requires_grad=True)
W2 = torch.randn(3, 64, requires_grad=True)
b2 = torch.randn(3, requires_grad=True)

parameters = [W1, b1, W2, b2]

# Forward pass
def forward(x, parameters):
    W1, b1, W2, b2 = parameters
    x = x @ W1.T + b1
    x = torch.relu(x)
    x = x @ W2.T + b2
    return x
```

**Problems:**
- 😫 Manual weight management
- 😫 Hard to add/remove layers
- 😫 Lots of boilerplate
- 😫 Easy to make mistakes

---

### **nn.Module Way (Professional):**

```python
class SpeakerCounter(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(5, 64)
        self.layer2 = nn.Linear(64, 3)
    
    def forward(self, x):
        x = self.layer1(x)
        x = torch.relu(x)
        x = self.layer2(x)
        return x

# Create model
model = SpeakerCounter()
```

**Benefits:**
- ✅ Automatic parameter tracking
- ✅ Easy to modify
- ✅ Clean and readable
- ✅ Industry standard

## **🧱 COMMON nn.Module LAYERS**

### **1. nn.Linear (Fully Connected Layer)**

```python
layer = nn.Linear(input_size, output_size)

# Replaces:
# W = torch.randn(output_size, input_size, requires_grad=True)
# b = torch.randn(output_size, requires_grad=True)
# output = input @ W.T + b
```

**Usage:**
```python
layer = nn.Linear(5, 64)  # 5 inputs → 64 outputs
x = torch.randn(10, 5)     # Batch of 10 samples
output = layer(x)          # Shape: (10, 64)
```

### **2. Activation Functions**

```python
# Option 1: Functional (like you've been using)
x = torch.relu(x)
x = torch.sigmoid(x)
x = torch.tanh(x)

# Option 2: Module (can be added to Sequential)
self.relu = nn.ReLU()
x = self.relu(x)
```

### **3. nn.Sequential (Stack Layers)**

```python
# Instead of defining each layer separately:
self.layer1 = nn.Linear(5, 64)
self.relu1 = nn.ReLU()
self.layer2 = nn.Linear(64, 32)
self.relu2 = nn.ReLU()
self.layer3 = nn.Linear(32, 3)

# Use Sequential:
self.network = nn.Sequential(
    nn.Linear(5, 64),
    nn.ReLU(),
    nn.Linear(64, 32),
    nn.ReLU(),
    nn.Linear(32, 3)
)
```

### **4. nn.Dropout (Regularization)**

```python
self.dropout = nn.Dropout(p=0.5)  # Drop 50% of neurons

# In forward:
x = self.dropout(x)  # Only active during training
```

### **5. nn.BatchNorm1d (Normalization)**

```python
self.bn = nn.BatchNorm1d(64)  # For 64 features

# In forward:
x = self.layer1(x)
x = self.bn(x)       # Normalize
x = torch.relu(x)
```

## TODO 3: Rebuild Your Speaker Counter

````python
def feature_extraction(
        audio: torch.Tensor
) -> torch.Tensor:
    mean = torch.mean(audio)
    std = torch.std(audio)
    min = torch.min(audio)
    max = torch.max(audio)
    square = audio ** 2
    energy = torch.mean(square)
    
    return torch.stack([mean, std, min, max, energy])
````

````python
def load_training_batch(
    manifest_path: Path,
    num_samples: int = 5,
    device: str = 'cpu'
) -> Tuple[torch.Tensor, torch.Tensor, torch.Tensor, List[Dict]]:
    
    # TODO 1: Load manifest JSON
    # Hint: json.load()
    with open(manifest_path, 'r') as f:
        manifest = json.load(f)
    
    # Take only num_samples entries
    manifest = random.sample(manifest, num_samples)
    
    # Initialize lists
    mixture_tensors = []
    label_tensors = []
    
    # TODO 2: Loop through each entry in manifest
    for entry in manifest:
        # TODO 2.1: Load mixture audio
        mixture_path = Path(entry['mixture_path'])
        mixture_audio, _ = load_audio(mixture_path)  # Use load_audio
        
        # TODO 2.2: Convert to tensor and pad/trim
        mixture_tensor = torch.from_numpy(mixture_audio)
        mixture_tensor = mixture_tensor.view(1, 1, -1)  # (1, 1, samples)
        mixture_tensor = feature_extraction(mixture_tensor)
        mixture_tensors.append(mixture_tensor)
        
        
        # TODO 2.6: Convert labels to tensor
        # Shape is (num_frames, num_speakers)
        # We need (max_speakers, num_frames) with padding
        speaker_count = int(entry['num_speakers']) - 1
        label_tensor = torch.tensor(speaker_count, dtype=torch.long)
        
        label_tensors.append(label_tensor)
        
    
    # TODO 3: Stack all tensors into batches
    mixtures_batch = torch.stack(tensors=mixture_tensors, dim=0).to(device)  # torch.cat along dim=0
    labels_batch = torch.stack(label_tensors).to(device)    # torch.stack
    
    return mixtures_batch, labels_batch
````

````python
class SpeakerCounter(nn.Module):
    """
    Classifier that predicts number of speakers (1, 2, or 3)
    from audio features.
    """
    def __init__(self, input_size=5, hidden_sizes=[64, 32, 16], num_classes=3):
        super().__init__()
        # TODO: Build network with given architecture
        # Hint: Use nn.Sequential or define layers individually
        self.network = nn.Sequential(
            nn.Linear(input_size, hidden_sizes[0]),
            nn.BatchNorm1d(hidden_sizes[0]),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(hidden_sizes[0],hidden_sizes[1]),
            nn.BatchNorm1d(hidden_sizes[1]),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(hidden_sizes[1],hidden_sizes[2]),
            nn.BatchNorm1d(hidden_sizes[2]),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(hidden_sizes[2],num_classes)
        )
        
    def forward(self, x):
        # TODO: Implement forward pass
        return self.network(x)
````

## TODO 4: Training with nn.Module

````python
def compute_accuracy(predictions, labels):
    """
    predictions: (batch_size, 3) logits
    labels: (batch_size,) true classes
    
    Returns: accuracy as percentage
    """
    predicted_classes = torch.argmax(predictions, dim=1)
    correct = (predicted_classes == labels).sum().item()
    total = labels.size(0)
    return 100.0 * correct / total
````

````python
TOTAL_SAMPLES = 10000 
batch_size = 64
steps_per_epoch = TOTAL_SAMPLES // batch_size

num_epochs = 5

# Create model
model = SpeakerCounter().to(device)

# Create optimizer (replaces your SimpleOptimizer!)
optimizer = optim.Adam(model.parameters(), lr=0.0001)

# Training loop
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    running_accuracy = 0.0
    
    for step in range(steps_per_epoch):
        # Load batch
        features, labels = load_training_batch(
            train_manifest_path, 
            batch_size, 
            device=device
        )
        
        # Forward pass
        logits = model(features)  # ← Much cleaner!
        
        # Compute loss
        loss = F.cross_entropy(logits, labels)
        
        # Backward pass
        loss.backward()
        
        # Update weights
        optimizer.step()
        optimizer.zero_grad()
        
        # Compute accuracy
        running_loss += loss.item()
        running_accuracy += compute_accuracy(logits, labels)
        
        # Optional: Print batch progress every 50 batches within the epoch
        if (step + 1) % 50 == 0:
            print(f"  Batch {step+1}/{steps_per_epoch} | Loss: {loss.item():.4f}")
    
    # 3. Print Summary at the end of each FULL epoch
    avg_epoch_loss = running_loss / steps_per_epoch
    avg_epoch_acc = running_accuracy / steps_per_epoch
    print(f"--- Epoch {epoch+1}/{num_epochs} Summary ---")
    print(f"Average Loss: {avg_epoch_loss:.4f} | Average Accuracy: {avg_epoch_acc:.2f}%\n")
````
